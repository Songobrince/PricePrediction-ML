```python
import random

max_price = 2872
optimal_altitude = 2130
min_altitude = 1980
max_altitude = 2500
constant = optimal_altitude / max_altitude

# Custom function to calculate altitude
def calculate_altitude(price):
    threshold = price / max_price
    altitude = max_altitude * threshold * constant
    return altitude

def calculate_temperature(price):
    threshold = price / max_price
    temperature  = 35 * threshold 
    return temperature

def calculate_humidity(price):
    threshold = price / max_price
    humidity = 80 * threshold 
    return humidity


def calculate_rainfall(price):
    threshold = price / max_price
    altitude = 1226 * threshold
    return altitude

def calculate_pyrethrum_content(price):
    threshold = price / max_price
    PC  = 2.1 * threshold
    return PC

def calculate_pyrethrum_content(price):
    threshold = price / max_price
    PC  = 2.1 * threshold
    return PC

def calculate_pH(price):
    threshold = price / max_price
    PC  = threshold
    return PC


# Apply the function to create a new column 'Altitude'
pyrethrum_data['Altitude'] = pyrethrum_data['Price'].apply(calculate_altitude)
pyrethrum_data['Rainfall'] = pyrethrum_data['Price'].apply(calculate_rainfall)
pyrethrum_data['Pyrethrin_Content'] = pyrethrum_data['Price'].apply(calculate_pyrethrum_content)
pyrethrum_data['Temperature'] = pyrethrum_data['Price'].apply(calculate_temperature)
pyrethrum_data['Humidity'] = pyrethrum_data['Price'].apply(calculate_humidity)

pyrethrum_data.head(-50)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
x = newdata.drop('Price', axis=1)
y = newdata['Price']  # Use square brackets to access the 'Price' column
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
# // create model  from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LinearRegression

# Create a Linear Regression model
model = LinearRegression()

# Fit the model
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Make predictions on the test data
y_pred = model.predict(x_test)


# Evaluate the model
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("R-squared:", r2)
```

```python
# // save the model 
import pickle
with open('pyrethrum_prediction_price.pkl', 'wb') as files:
    pickle.dump(model,files)
```
